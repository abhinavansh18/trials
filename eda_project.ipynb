```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
```

```python
#need to clean the data and manage the missing values
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.head()
```

```python
#type conversion of column
df["Region"]=df["Region"].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#remove null
df=df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
#we have to convert all other columns to float as it belongs to object type
objects=[features for features in df.columns if df[features].dtype=='O']
objects
```

```python
for i in objects:
  if i!='Classes':
    df[i]=df[i].astype(float)
```

```python
df.to_csv("Algerian_forest_fires_dataset_clean.csv",index=False)
```
